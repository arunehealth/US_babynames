####Finding the most common names by popularity 

```python
names_dict = dict()
for index, row in n_n.iterrows():
    if row['Name'] not in names_dict:
        names_dict[row['Name']] = row['Count']
    else:
            names_dict[row['Name']] += row['Count'] 
```

```python
top_10 = Counter(names_dict).most_common (10)
print ("Top 10 names = ")
for pair in top_10:
    print (pair[0])
```
